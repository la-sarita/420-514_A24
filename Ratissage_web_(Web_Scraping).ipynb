# Web Scraping (toscrape.com)

```python
import requests
from bs4 import BeautifulSoup

# Requête pour obtenir la page web
#url = "https://books.toscrape.com/"
# Avec pgination
global num_page
url = f'https://books.toscrape.com/catalogue/page-{num_page}.html'

Max_PAGE = 50
HEADERS = ({'User-Agent':
'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko)'
'Chrome/44.0.2403.157 Safari/537.36',
'Accept-Language': 'fr-CH, fr;q=0.9'})

def getNextUrl():
    return f'https://books.toscrape.com/catalogue/page-{num_page}.html'
```

```python
def extract_book(article):
    title = article.find('h3').find('a').get('title').strip()
    price = article.find('p', class_='price_color').get_text().strip()
    rating = article.find('p', class_='star-rating').get('class')[1]
    stock = article.find('p', class_='instock availability').get_text().strip()
    # Méthode 1
    content = {
        'title': title,
        'price': price,
        'rating': rating,
        'stock': stock
    }
    list_articles.append(content)
    # Méthode 2
    return [title, price, rating, stock]  # book info

def extraxt_books(soup):
    books = []
    num_page = 1
    while num_page <= Max_PAGE:
      url = getNextUrl()
      articles = soup.find_all('article')
      for article in articles:
          books.append(extract_book(article))
      print(num_page, ' / ', Max_PAGE)
      num_page += 1
    return books

```

```python
response = requests.get(url, headers=HEADERS)

# Vérification du statut de la réponse
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    #print(soup.prettify())
    # Affichage des titres
    list_articles = []
    books = extraxt_books(soup)
else:
    print("Erreur de chargement de la page")
```

```python
import pandas as pd
import numpy as np

columns = ['title', 'price', 'rating', 'stock']
data = pd.DataFrame(columns=columns, data=np.array(books))
data.head(50)
```

```python
def isfloat(value):
  try:
    float(value)
    return True
  except ValueError:
    return False

def transformer_float(x):
    if x :
        if isfloat(x):
            return round(float(x), 2)
        else :
            return None
```

```python
data.price = data.price.apply(lambda x : x.replace('£', ''))
data.price = data.price.apply(transformer_float)
data
```
